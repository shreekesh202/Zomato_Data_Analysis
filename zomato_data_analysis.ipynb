# Checking the distribution of each column

```python
num_col=df_1.select_dtypes(include=['int','float']).columns
num_col
```

```python
for col in num_col:
    if df_1[col].dtype=='int' or df_1[col].dtype=='float':
        sns.histplot(data=df_1,x=df_1[col],kde=True,bins=10)
        plt.ylabel('counts')
        plt.title(f'Distribution of {col}')
        plt.show()
```

```python
for col in num_col:
    plt.figure(figsize=(5,1))
    sns.boxplot(data=df_1,x=df_1[col])
    plt.title(f'Boxplot for {col}')
    plt.show()
```

# Visualization of Prices & Volume over Time

```python
plt.figure(figsize=(15,8))
sns.lineplot(data=df_1,x='Date',y='High',label='High')
sns.lineplot(data=df_1,x='Date',y='Low',label='Low')
plt.legend()
plt.ylabel('Price')
plt.xlabel('Date')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(data=df_1,x='Date',y='Open',label='Open')
sns.lineplot(data=df_1,x='Date',y='Close',label='Close')
plt.ylabel('Price')
plt.xlabel('Date')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(data=df_1,x='Date',y='Volume')
plt.ylabel('Volume')
plt.xlabel('Year')
plt.show()
```
